#### Retrieving samples

#### Importing Perseus non-transformed matrix

```python
matrix_perseus = pd.read_perseus(r"C:\Users\denise\Sources\Repos\MSC_THESIS_PUBLIC\data\quantile_data\matrix_27_notrans_limma.txt")
sample_perseus = matrix_perseus.filter(regex='_NB_Plasma_Quant_C18_Pos_')
sample_perseus
```

#### Transforming matrix using log2(x)

```python
sample_log_custom = sample_perseus.apply(lambda x: numpy.log2(x))
sample_log_custom = sample_log_custom.round(decimals= 4)
```

```python
def quantile_normalization(dataframe_input):
    df = dataframe_input.copy()
    dic = {}                                                                                         # dictionary 
    for column in df:
        dic.update({column : sorted(df[column])})                                                    # key : ordered value
    sorted_df = pd.DataFrame(dic) 
    rank = sorted_df.mean(axis = 1).tolist()                                                            #rank is a list of mean along column
    for column in df:
        t = numpy.searchsorted(numpy.sort(df[column]), df[column])                                       # t is the list per column of ordered values of original df
        df[column] = [rank[i] for i in t]                                                                  # rank replace the position in t
    return df
```

#### Normalizing samples with quantile normalization

```python
sample_quantile_custom = quantile_normalization(sample_log_custom)
sample_quantile_custom = sample_quantile_custom.round(decimals = 4)
sample_quantile_custom
```

#### Importing Perseus quantile-normalized matrix

```python
matrix_quantile_perseus = pd.read_perseus(r"C:\Users\denise\Sources\Repos\MSC_THESIS_PUBLIC\data\quantile_data\matrix30_quantile_limma.txt")
sample_quantile_perseus = matrix_quantile_perseus.filter(regex='_NB_Plasma_Quant_C18_Pos_')
sample_quantile_perseus = sample_quantile_perseus.round(decimals = 4)
matrix_quantile_perseus
```

#### Computing difference between Perseus' quantile-normalized matrix and our quantile-normalized matrix

##### (True = equal values, False = different values)

```python
exactly_equal = pd.DataFrame.eq(sample_quantile_perseus, sample_quantile_custom)
exactly_equal
```

#### Counting how much False values we have

```python
all_false = exactly_equal[exactly_equal==False].count(axis=0)
all_false = numpy.array(all_false)
False_value = pd.DataFrame()
False_value['num for each col'] = all_false
False_value['percentage'] = [(x*100) /len(exactly_equal) for x in all_false ]
False_value
```

```python
mean_false_value = numpy.mean(all_false)
variance_false_value = numpy .var(all_false)
print(mean_false_value)
print(variance_false_value)
```
